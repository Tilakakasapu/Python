```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure,output_file,show
```

```python
#we can create a bokeh plot using python code
x = [1,2,3,4,5]
y=[5,6,3,6,4]
output_file('test.html')
p = figure(title = 'bokeh plot')
p.circle(x,y)
show(p)
```

```python
#glyphs refer to visual shapes and markers that are used to represent data points on a plot.
x = [1,2,3,4,5]
y=[5,6,3,6,4]
output_file('test.html')
p = figure(title = 'bokeh plot')
p.circle(x,y)
show(p)
```

```python
#in the above example the circle glyph is used to creaate circles at speccified coordinates representing the data points
#glyphs can be circle , line , rectangle,bers,triangles
```

```python
output_file('test.html')
p = figure(title = 'bokeh plot')#from here the title can be customized
p.circle(x,y,legend_label= 'bokeh_label',color = 'red')#from here the color and the legend can be controlled
show(p)
```
