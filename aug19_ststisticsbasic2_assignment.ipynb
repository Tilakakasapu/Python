Q1. **Three Measures of Central Tendency:**
The three measures of central tendency are:
- Mean
- Median
- Mode

Q2. **Difference between Mean, Median, and Mode:**
   - **Mean:** The mean is the average of all values in a dataset. It's obtained by summing up all the values and dividing by the number of values.
   - **Median:** The median is the middle value when the data is ordered. It's not affected by extreme values and provides a measure of the central value.
   - **Mode:** The mode is the most frequently occurring value in a dataset. There can be one mode (unimodal), multiple modes (multimodal), or no mode.

```python
import numpy as np

data = [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]
std_dev = np.std(data)
print("Standard Deviation:", std_dev)

Q5. **Measures of Dispersion:**
   - **Range:** Difference between the maximum and minimum values.
   - **Variance:** Average of squared differences from the mean.
   - **Standard Deviation:** Square root of the variance.

These measures quantify how spread out the data values are from the central value. For example, a high standard deviation indicates a wide spread of values, while a low standard deviation suggests values are close to the mean.

Q6. **Venn Diagram:**
A Venn diagram is a graphical representation that uses circles to show the relationships between sets. It illustrates the common and unique elements among sets.

Q7. **Set Operations for A and B:**
(i) A ∩ B (intersection): The elements that are common to both sets A and B = {2, 6}
(ii) A ⋃ B (union): The elements in either set A or set B (including duplicates) = {0, 2, 3, 4, 5, 6, 7, 8, 10}

Q8. **Skewness in Data:**
Skewness measures the asymmetry of a probability distribution. In data, it indicates whether the distribution is skewed to the left (negatively skewed), balanced (symmetrical), or skewed to the right (positively skewed).

Q9. **Median Position in Right Skewed Data:**
In a right-skewed distribution, the median will be less than the mean. This is because the long tail of the distribution pulls the mean in the direction of the skewness.

Q10. **Covariance vs. Correlation:**
   - **Covariance:** Measures the direction of the linear relationship between two variables. It can be positive (both variables increase together), negative (one variable increases while the other decreases), or close to zero (no linear relationship).
   - **Correlation:** Measures both the strength and direction of the linear relationship between two variables. It's always between -1 and 1, where -1 indicates a perfect negative linear relationship, 1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship.

These measures help analyze the relationship between variables and understand how changes in one variable relate to changes in another.

Q11. **Formula for Sample Mean:**
Sample Mean = Sum of all values / Number of values

Example: Dataset = [10, 15, 20, 25, 30]
Sample Mean = (10 + 15 + 20 + 25 + 30) / 5 = 20

Q12. **Relationship between Measures of Central Tendency in Normal Distribution:**
In a normal distribution, the mean, median, and mode are approximately equal, creating a symmetric bell-shaped curve.

Q13. **Difference between Covariance and Correlation:**
Covariance measures the extent to which two variables change together, but it doesn't provide the strength or direction of the relationship. Correlation, on the other hand, standardizes the covariance to always be between -1 and 1, providing information about both the strength and direction of the linear relationship.

Q14. **Outliers and Measures of Central Tendency/Dispersion:**
- Outliers can significantly affect the mean, pulling it towards the outlier's value.
- Median is less affected by outliers, making it a robust measure of central tendency.
- Measures of dispersion like range, variance, and standard deviation can increase significantly due to outliers, indicating greater spread.

Example: If you have a dataset of exam scores with an outlier of 1000, the mean would be heavily influenced by the outlier, leading to a skewed interpretation of the central tendency. Additionally, the standard deviation would be inflated, indicating higher dispersion due to the outlier's large distance from the mean.
