```python
#matlplotlib is a python library used for visualising the data in charts,bargraphs and histograms
#it is used when there is a large amount of data and after ploting the data we can understand the data and how it increases and decreases
#plots : bar graph,histogram,scatter points,lineplots,piecharts
```

```python
#scatter plot is the plot of all the data points in the x-y plane 
#it contains points
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel('x-axis')
plt.ylabel('ylabel')
plt.title('title')
plt.plot(x,y)
```

```python
#subplot is used to create multiple plots in a single figure it is mainly useful when you want to create multiple plots side by side 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('plot 1')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('plot 2')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('plot 3')
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('plot 4')
```

```python
#BAR PLOT is used to create bar graph using the data it is mainly use ful when we are creating plot when there is categorical data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

```python
#box plot is a statistical visualisation that 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
```
